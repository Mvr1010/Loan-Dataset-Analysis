## Introduction

In the world of finance, loans are a crucial component that allows individuals, businesses, and organizations to access capital for various purposes. A loan can be defined as a financial agreement between a lender and a borrower, where the borrower receives a certain amount of money that must be repaid with interest within a specified period.

Loans datasets are collection of data points that have been gathered with the objective of identifying key variables that can help predict whether or not an individual is likely to default on a loan. Loans datasets are particularly valuable for financial institutions as it provides them with insights that can be used to identify risky loan applicants and avoid any potential financial losses. By analysing the data contained within the loans datasets, financial companies can make more informed lending decisions, reduce their risk exposure, and ensure that their lending practices are both responsible and sustainable. Ultimately, the loans datasets play a crucial role in helping financial institutions to make better decisions and manage their lending operations more effectively.

Loans datasets are collection of data related to loans, including borrower information, loan amount, interest rates, loan term, repayment schedules, and other relevant details. These datasets are often used by financial institutions, credit rating agencies, and researchers to analyse lending trends, risk assessment, and other important factors related to loans.

## Data Dictionsary

| No.  | Column Name | Description |
| :-- | :-- | :-- |
| 1 | annual_inc | The self-reported annual income provided by the borrower during registration |
| 2 | dti | A ratio calculated using the borrower’s total monthly debt payments on the total debt obligations, excluding mortgage and the requested LC loan, divided by the borrower’s self-reported monthly income. |
| 3 | emp_length | Employment length in years. Possible values are between 0 and 10 where 0 means less than one year and 10 means ten or more years. |
| 4 | grade | LC assigned loan grade |
| 5 | funded_amnt | The total amount committed to that loan at that point in time. |
| 6 | funded_amnt_inv | The total amount committed by investors for that loan at that point in time. |
| 7 | id | A unique LC assigned ID for the loan listing. |
| 8 | installment | The monthly payment owed by the borrower if the loan originates. |
| 9 | last_pymnt_amnt | Last total payment amount received |
| 10 | int_rate | Interest Rate on the loan |
| 11 | last_pymnt_d | Last month payment was received |
| 12 | loan_amnt | The listed amount of the loan applied for by the borrower. If at some point in time, the credit department reduces the loan amount, then it will be reflected in this value. |
| 13 | loan_status | Current status of the loan |
| 14 | member_id | A unique LC assigned Id for the borrower member. |
| 15 | purpose | A category provided by the borrower for the loan request. |
| 16 | term | The number of payments on the loan. Values are in months and can be either 36 or 60. |
| 17 | total_acc | The total number of credit lines currently in the borrower's credit file |
| 18 | total_pymnt | Payments received to date for total amount funded |
| 19 | total_pymnt_inv | Payments received to date for portion of total amount funded by investors |
| 20 | total_rec_int | Interest received to date |

#### 11.Using a user defined function convert the ‘emp_len’ column into categorical column as follows - 
#### If emp_len is less than equals to 1 then recode as ‘fresher’. If emp_len is greater than 1 and less than 3 then recode as ‘junior’. If emp_len is greater than 3 and less than 7 then recode as ‘senior’. If emp_len is greater than 7 then recode as ‘expert’.

```python
# creating an user defined function to convert numeric value to categorical value based on given condition

def convert(x):
    '''
    this function is used for converting a numeric value to categorical value based on given conditions
    x takes the values of emp_length_num
    '''
    
    if x <= 1: 
        return 'Fresher'         # returning 'fresher' when emp_length_num <=1
    elif (x > 1) and (x <= 3):
        return 'Junior'          # returning 'junior' when emp_length_num >1 and <3
    elif (x > 3) and (x <= 7):
        return 'Senior'          # returning 'senior' when emp_length_num >3 and <7
    else:
        return 'Expert'          # returning 'expert' when emp_length_num >7
```

```python
# creating a new column emp_experience and using .map to convert each value from emp_length_num column to a categorical value 
# by calling the user defined function convert()

loan["emp_experience"]= loan.emp_length_num.map(convert) 

loan
```
